```python
import pandas as pd
import pandera as pa
```

```python
valores_ausentes = ['###!','####','**','****','*****','NULL']
df = pd.read_csv("ocorrencia_2010_2020.csv", parse_dates=['ocorrencia_dia'], dayfirst=True, na_values = valores_ausentes) ##alterando os não informados
df.head(10)
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo":pa.Column(pa.Int, required=False), ##caso a coluna seja opcional
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String, nullable=True), ##Possui null
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo": pa.Column(pa.String, nullable=True),
        "ocorrencia_dia": pa.Column(pa.DateTime, nullable=True),
        "ocorrencia_hora": pa.Column(pa.String,pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Float, nullable=True)
    }
)
```

```python
schema.validate(df)
```

```python
#ocorrencias com classificação == "INCIDENTE GRAVE"
filtro = df.ocorrencia_classificacao == "INCIDENTE GRAVE"
df.loc[filtro]
```

```python
#cidades com ocorrencias com classificação == "INCIDENTE GRAVE" no estado "SP"
filtro1 = df.ocorrencia_classificacao == "INCIDENTE GRAVE"
filtro2 = df.ocorrencia_uf == "SP"
df.loc[filtro1 & filtro2]
```

```python
#cidades com ocorrencias com classificação == "INCIDENTE GRAVE" ou estado "SP"
filtro1 = df.ocorrencia_classificacao == "INCIDENTE GRAVE"
filtro2 = df.ocorrencia_uf == "SP"
df.loc[filtro1 | filtro2]
```

```python
#cidades com ocorrencias com classificação == "INCIDENTE GRAVE" ou classificação == "INCIDENTE" no estado "SP"
filtro1 = (df.ocorrencia_classificacao == "INCIDENTE GRAVE") | (df.ocorrencia_classificacao == "INCIDENTE") #() para mostrar a sequencia da análise
filtro2 = df.ocorrencia_uf == "SP"
df.loc[filtro1 & filtro2]
```

```python
#cidades com ocorrencias com classificação == "INCIDENTE GRAVE" ou classificação == "INCIDENTE" no estado "SP"
filtro1 = df.ocorrencia_classificacao.isin (['INCIDENTE GRAVE','INCIDENTE']) #isin busca na lista ao inves de colocar vários "or"
filtro2 = df.ocorrencia_uf == "SP"
df.loc[filtro1 & filtro2]
```

```python
#ocorrencias cuja a cidade comece com "c"
filtro = df.ocorrencia_cidade.str[0] == 'C'
df.loc[filtro]
```

```python
#ocorrencias cuja a cidade terminam com "a"
filtro = df.ocorrencia_cidade.str[-1] == 'A'
df.loc[filtro]
```

```python
#ocorrencias cuja a cidade terminam com "ma"
filtro = df.ocorrencia_cidade.str[-2:] == 'MA'
df.loc[filtro]
```

```python
#criando coluna data e hora juntos
df['ocorrencia_dia_hora'] = pd.to_datetime(df.ocorrencia_dia.astype(str) + ' ' + df.ocorrencia_hora) #transformou ocorrencia_dia em string, concateou e transformou em data
```

```python
df.fillna(value={'total_recomendacoes':'0'}, inplace=True)
```

```python
df['total_recomendacoes'] = df.total_recomendacoes.astype(int)
```

```python
#ocorrencias de 2015 e mês 12 e do dia 3 ao 8
filtro_ano = df.ocorrencia_dia_hora.dt.year == 2015
filtro_mes = df.ocorrencia_dia_hora.dt.month == 12
filtro_dia_inicio = df.ocorrencia_dia_hora.dt.day > 2
filtro_dia_fim =df.ocorrencia_dia_hora.dt.day < 9
df.loc[filtro_ano & filtro_mes & filtro_dia]
```

```python
#ocorrencias de 2015 e mês 12 e do dia 3 ao 8 das 11 as 13 horas
filtro_hora_inicio = df.ocorrencia_dia_hora >= '2015-12-03 11:00:00'
filtro_hora_fim =  df.ocorrencia_dia_hora <= '2015-12-08 13:00:00'
df.loc[filtro_hora_inicio & filtro_hora_fim]
```
